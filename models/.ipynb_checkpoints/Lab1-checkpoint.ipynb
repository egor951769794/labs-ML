```python
import random
import numpy as np

class LinearRegression:
    def __init__(self, X, y, test_count : int, no_iters: int):
        self.X = X
        self.y = y
        self.test_count = test_count
        self.no_iters = no_iters
        
        self.X_test = X.iloc[:test_count:]
        self.X_train = X.iloc[test_count:,:]
        
        self.y_test = y.iloc[:test_count]
        self.y_train = y.iloc[test_count:]
        
        self.sample_count, self.sample_length = self.X_train.shape
        
        self.w = np.zeros(self.sample_length)
        
    def fit(self):
#         for i in range(self.no_iters):
        L = (1 / self.sample_count)*np.square(self.y_train - np.dot(self.X_train, self.w))
        print(L)
        
```

```python
lr = LinearRegression(X, y, 10000, 10)
```
