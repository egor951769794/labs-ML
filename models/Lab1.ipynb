```python
class linReg:
    def __init__(self, lr=0.001, iters=100, eps=1, alpha=0.5):
        self.lr = lr
        self.iters = iters
        self.eps = eps
        self.alpha = alpha
  
    def fit(self, x, y):
        d = np.inf
        X = self.transform(x)
        
        w = np.zeros(X.shape[1])
        iter = 0

        while d > self.eps and iter <= self.iters:
            loss = self.get_loss(X, y, w, self.alpha)
            w = w - (self.lr * 2 * np.dot(X.T, np.dot(X, w) - y) / X.shape[0] + 2 * self.alpha * w / X.shape[0])
            d = np.abs(loss - self.get_loss(X, y, w, self.alpha))
            iter += 1
            print(d)

        self.w = w

    def transform(self, x):
        return np.concatenate((x,np.ones((len(x), 1))), axis=1)

    def get_loss(self, x, y, w, hyperalpha):
        return sum((y - np.dot(x, w)) ** 2) / x.shape[0] + hyperalpha * sum(np.square(w)) / x.shape[0]
        
    def MAE(self, x, y):
        result = 0
        x = self.transform(x)
        y_pred = np.dot(x, self.w)
        for i in range(x.shape[0]):
            result += abs(y.iloc[i] - y_pred[i])
        return result / x.shape[0]
    
    def MSE(self, x, y):
        result = 0
        x = self.transform(x)
        y_pred = np.dot(x, self.w)
        for i in range(x.shape[0]):
            result += (y.iloc[i] - y_pred[i])**2
        return result / x.shape[0]
    
    def RMSE(self, x, y):
        result = 0
        x = self.transform(x)
        y_pred = np.dot(x, self.w)
        for i in range(x.shape[0]):
            result += (y.iloc[i] - y_pred[i])**2
        return (result / x.shape[0])**0.5
    
    def MAPE(self, x, y):
        result = 0
        x = self.transform(x)
        y_pred = np.dot(x, self.w)
        for i in range(x.shape[0]):
            result += abs((y.iloc[i] - y_pred[i]) / y.iloc[i])
        return result / x.shape[0]
    
    def R2(self, x, y):
        mse_result = 0
        x = self.transform(x)
        y_pred = np.dot(x, self.w)
        
        y_avg = sum(y) / y.shape[0]
        
        for i in range(x.shape[0]):
            mse_result += (y.iloc[i] - y_pred[i])**2
        mse_result /= x.shape[0]
        
        denominator = 0
        for i in range(x.shape[0]):
            denominator += (y.iloc[i] - y_avg)**2
        denominator /= x.shape[0]
            
        return (1 - mse_result / denominator)
    
```

```python
lr2 = linReg(lr=0.000000000001, iters=250)

X1 = lr1.X_train.copy(deep=True)
Y1 = lr1.y_train.copy(deep=True)

X1_test = lr1.X_test.copy(deep=True)
Y1_test = lr1.y_test.copy(deep=True)

lr2.fit(X1, Y1)
```

```python
print("MAE:", lr2.MAE(X1_test, Y1_test))
print("MSE:", lr2.MSE(X1_test, Y1_test))
print("RMSE:", lr2.RMSE(X1_test, Y1_test))
print("MAPE:", lr2.MAPE(X1_test, Y1_test))
print("R_squared:", lr2.R2(X1_test, Y1_test))
```
